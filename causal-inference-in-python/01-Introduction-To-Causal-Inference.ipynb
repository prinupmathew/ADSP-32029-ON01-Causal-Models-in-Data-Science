# 01 - Introduction To Causal Inference



## What is Causal Inference



Causal inference is the set of methods used to estimate cause-and-effect relationships from data—answering questions like "What would happen if we changed X?" rather than just "Are X and Y associated?" It relies on assumptions and study design (experiments or quasi-experiments) to distinguish correlation from causation and to estimate counterfactual outcomes.

## Why we Do Causal Inference

We do causal inference to predict the effect of actions, not just observe patterns. It helps answer policy and decision questions like whether a treatment, price change, or campaign will change outcomes, and by how much. Causal methods also separate confounding from true effects, improve fairness and generalization, and guide interventions that would not be valid from correlations alone.

## Machine Learning and Causal Inference

**Goal**
- Machine learning: predict $Y$ accurately from features $X$.
- Causal inference: estimate the effect of changing $T$ (treatment) on $Y$.

**Question**
- Machine learning: "Given $X$, what is $Y$?"
- Causal inference: "If we set $T=1$ instead of $T=0$, how does $Y$ change?"

**Example**
- ML: Predict weekly sales from price, season, and promotion history.
- Causal: Estimate the sales lift caused by running a promotion, controlling for seasonality and baseline demand.

### Causal Quantities: An Example
 

```python
pd.DataFrame(dict(
    i= [1,2,3,4,5,6],
    y0=[200,120,300, 450,600,600],
    y1=[220,140,400, 500,600,800],
    t= [0,0,0,1,1,1],
    x= [0,0,1,0,0,1],
)).assign(
    y = lambda d: (d["t"]*d["y1"] + (1-d["t"])*d["y0"]).astype(int),
    te=lambda d: d["y1"] - d["y0"]
)
```

```python
pd.DataFrame(dict(
    i= [1,2,3,4,5,6],
    y0=[200,120,300, np.nan, np.nan, np.nan,],
    y1=[np.nan, np.nan, np.nan, 500,600,800],
    t= [0,0,0,1,1,1],
    x= [0,0,1,0,0,1],
)).assign(
    y = lambda d: np.where(d["t"]==1, d["y1"], d["y0"]).astype(int),
    te=lambda d: d["y1"] - d["y0"]
)
```
